### Build inverted index

### Raw code

Add score 1 if the word exists in the opposite database

```python
score = {}
for index_x,x in enumerate(text_base):
    print(index_x)
    score[index_x] = {}
    split_words = x.split()
    for index_y,y in enumerate(text_var):
        score[index_x][index_y] = 0
        for w in split_words:
            try:
                if bool(re.search(w, y)):
                    score[index_x][index_y] += 1
            except:
                continue
        break
```
